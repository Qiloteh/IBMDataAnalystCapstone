### 1. Load the Dataset


<h5>1.1 Import necessary libraries and load the dataset.</h5>


Ensure the dataset is loaded correctly by displaying the first few rows.


```python
# Import necessary libraries
import pandas as pd

# Load the Stack Overflow survey data
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# Display the first few rows
print(df.head())
```

#### 2. Explore the Dataset


<h5>2.1 Summarize the dataset by displaying the column data types, counts, and missing values.</h5>


```python
# Write your code here
# Create a summary with data types, non-null counts, and missing values
summary = pd.DataFrame({
    "Data Type": df.dtypes,
    "Non-Null Count": df.count(),
    "Missing Values": df.isnull().sum()
})

print(summary)
```

<h5>2.2 Generate basic statistics for numerical columns.</h5>


```python
# Write your code here
df.describe()
```

### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
# Write your code here
# Step 1: Check unique values in the 'Country' column
unique_countries = df['Country'].unique()
print("Unique country values:\n", unique_countries)

# Step 2: Display value counts (to spot anomalies)
print("\nCountry value counts:\n")
print(df['Country'].value_counts(dropna=False))

# Step 3: Check for missing values
missing_countries = df[df['Country'].isnull()]
print("\nRows with missing Country values:\n", missing_countries)
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
## Write your code here
# Standardize country entries
df['Country'] = df['Country'].str.strip().str.title()

# Define a mapping dictionary
country_mapping = {
    'Usa': 'United States',
    'U.S.': 'United States',
    'United States Of America': 'United States',
    'Uk': 'United Kingdom',
    'U.K.': 'United Kingdom',
    'England': 'United Kingdom',
    'Uae': 'United Arab Emirates',
    'Republic Of India': 'India',
    'Narnia': pd.NA,
    'Earth': pd.NA,
}

# Apply the mapping
df['Country'] = df['Country'].replace(country_mapping)
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
## Write your code here
# Apply one-hot encoding to the Employment column
employment_encoded = pd.get_dummies(df['Employment'], prefix='Employment')

# Concatenate the new columns to the original DataFrame (and optionally drop the original column)
df = pd.concat([df.drop('Employment', axis=1), employment_encoded], axis=1)

# View the first few rows
print(df.head())
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
## Write your code here
# Count missing values per column
missing_values = df.isnull().sum()

# Sort in descending order to find the worst offenders
missing_sorted = missing_values[missing_values > 0].sort_values(ascending=False)

# Display the result
print("Columns with the highest number of missing values:\n")
print(missing_sorted)
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
## Write your code here
# Check for missing values in the column
print("Missing values before imputation:", df['ConvertedCompYearly'].isnull().sum())

# Impute with the mean
df['ConvertedCompYearly'] = df['ConvertedCompYearly'].fillna(df['ConvertedCompYearly'].mean())

print("Missing values after imputation:", df['ConvertedCompYearly'].isnull().sum())
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here
# Check how many missing values exist before imputation
print("Missing values before:", df['RemoteWork'].isnull().sum())

# Find the most frequent value (mode)
most_frequent = df['RemoteWork'].mode()[0]

# Fill missing values with the mode
df['RemoteWork'] = df['RemoteWork'].fillna(most_frequent)

# Confirm missing values have been imputed
print("Missing values after:", df['RemoteWork'].isnull().sum())
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
pip install -U scikit-learn
```

```python
## Write your code here

from sklearn.preprocessing import MinMaxScaler
# Ensure there are no missing values (you can impute them first if needed)
df['ConvertedCompYearly'] = df['ConvertedCompYearly'].fillna(df['ConvertedCompYearly'].median())

# Reshape data for sklearn (expects 2D array)
scaler = MinMaxScaler()
df['ConvertedCompYearly_Scaled'] = scaler.fit_transform(df[['ConvertedCompYearly']])

# View result
print(df[['ConvertedCompYearly', 'ConvertedCompYearly_Scaled']].head())
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
import numpy as np
```

```python
## Write your code here
# Impute missing values if necessary
df['ConvertedCompYearly'] = df['ConvertedCompYearly'].fillna(df['ConvertedCompYearly'].median())

# Filter out zero or negative values (log not defined for them)
df = df[df['ConvertedCompYearly'] > 0]

# Apply log transformation
df['ConvertedCompYearly_Log'] = np.log(df['ConvertedCompYearly'])

# View result
print(df[['ConvertedCompYearly', 'ConvertedCompYearly_Log']].head())
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
## Write your code here

# Handle string values in YearsCodePro
df['YearsCodePro'] = df['YearsCodePro'].replace({
    'Less than 1 year': 0,
    'More than 50 years': 51
})

# Convert to numeric (after replacing text values)
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Define experience levels
def categorize_experience(years):
    if pd.isnull(years):
        return 'Unknown'
    elif years < 3:
        return 'Beginner'
    elif 3 <= years < 7:
        return 'Intermediate'
    elif 7 <= years < 15:
        return 'Experienced'
    else:
        return 'Expert'

# Apply the function
df['ExperienceLevel'] = df['YearsCodePro'].apply(categorize_experience)

# Preview
print(df[['YearsCodePro', 'ExperienceLevel']].head())
```
